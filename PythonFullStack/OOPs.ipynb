```python
class BankAccount:

    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance   # private variable

    # private validation method
    def __validate_amount(self, amount):
        return amount > 0

    def deposit(self, amount):
        if self.__validate_amount(amount):
            self.__balance += amount
            print(f"‚Çπ{amount} deposited successfully")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if not self.__validate_amount(amount):
            print("Withdrawal amount must be positive")
        elif amount <= self.__balance:
            self.__balance -= amount
            print(f"‚Çπ{amount} withdrawn successfully")
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

    def display(self):
        print("Name:", self.name)
        print("Balance: ‚Çπ", self.__balance)


# Object creation
m1 = BankAccount("Robin", 5000)

# Method calls
m1.deposit(2000)
m1.withdraw(1500)
m1.display()
```

### BankMiniApp 
#### api call
#### db_call
#### security ->abstract method
#### WebApp-class

```python
from abc import ABC, abstractmethod

# Abstract Base Class
class BankMiniApp(ABC):

    @abstractmethod
    def security(self):
        pass

    @abstractmethod
    def api_call(self):
        pass

    @abstractmethod
    def db_call(self):
        pass


# Concrete Class
class WebApp(BankMiniApp):

    def security(self):
        print("Security: Username + Password + OTP")

    def api_call(self):
        print("API Call: Getting account details from server")

    def db_call(self):
        print("DB Call: Fetching data from database")


# Main execution
if __name__ == "__main__":
    app = WebApp()
    app.security()
    app.api_call()
    app.db_call()
```

```python
class SecureLocker:
    def __init__(self, pin):
       
        self.__pin = pin
        self.__locked = True

    
    def __verify_pin(self, pin):
        return self.__pin == pin

    
    def unlock(self, pin):
        if self.__verify_pin(pin):
            self.__locked = False
            print("Locker unlocked")
        else:
            print("Incorrect PIN")

    def lock(self):
        self.__locked = True
        print("Locker locked")

   
    def status(self):
        if self.__locked:
            print("Locker is Locked")
        else:
            print("Locker is Unlocked")




locker = SecureLocker(1234)

locker.status()        
locker.unlock(1111)   
locker.unlock(1234)    
locker.status()
locker.lock()
locker.status()
```

```python
# cannot access without super()
#method overriding
class Parent:
    def __init__(self, num):
        self.__num=num

    def get_num(self):
        return self.__num
class Child(Parent):
    def __init__(self,val, num):
        super().__init__(num)
        self.__val=val

    def get_val(self):
        return self.__val

son =Child(100,200)
print("Parent: Num:", son.get_num())
print("Child: Val:",son.get_val())

```

```python
#can super access parent ka data?
# using super outside the class
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
    def buy(self):
        print("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print("buying a smartphone")
        #syntax to call parent ka buy method
        super().buy()

s=SmartPhone(200, "Apple", 13)
s.buy()
```

```python
#Create a list that has all the methods of list 
#Print it one by one (iteration)

list = [11,22,33,44,55,66]
print(list)

#append()
list.append(77)
print(list)

#insert()
list.insert(6,88)

#remove()
list.remove(22)
print(list)

nums=[22,3,11,44,21]
#sort()
nums.sort()
print(nums)



```

```python
# List containing all list methods
list_methods = [
    "append()", "clear()", "copy()", "count()", "extend()",
    "index()", "insert()", "pop()", "remove()", "reverse()", "sort()"
]

# Printing methods one by one
for method in list_methods:
    print(method)
```

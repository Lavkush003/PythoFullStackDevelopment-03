```python
#Variable Length Arguments

#args=> many  positional args

def add(*nums):
    return sum(nums)
    print(type(nums))
a=add(1,2,3,4,5)

print(a)
print(type(a))
```

```python
# **kwars ==> many keyword argument
def profile(**data):
    print(type(data))
    print(data)
profile(name="Robin", age=25, sem ="6th")
```

```python
 #Anonymous Function (Lambda Function)
# A varaible is just a function without a name written in one line

add =lambda a,b:a+b
print(add(11,22))

sub =lambda x,y: x-y
print(sub(22, 11))

mul =lambda x,y:x*y
print(mul(10,10))

div = lambda x,y: x/y
print(div(20,2))

max_val =lambda a, b: a if a>b else b
print(max_val(10,20))

```
